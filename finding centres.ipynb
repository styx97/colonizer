```python
for i in range(a):
    for j in range(b):
        if (arena[i][j][0] or arena[i][j][1] or arena[i][j][2]) > 20:
            temp_circle = arena[i][j][:3]
            temp_circle = list(temp_circle)
            #print(temp_circle)    
        if temp_circle not in colors:
            #print(colors)
            d = i
            colors.append(temp_circle)
           
            while (arena[d][j][0] or arena[d][j][1] or arena[d][j][2]) > 20:
                d += 1
            radius = (d-i)//2
            centre = [i+radius,j]
            centres.append(centre)
            
            
```
